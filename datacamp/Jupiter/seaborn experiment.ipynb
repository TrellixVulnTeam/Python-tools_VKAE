```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

maindata = pd.read_csv("../datasets/pokemon.csv")
sns.regplot(x="Total", y="Defense", data=maindata )
print(maindata)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

maindata = pd.read_csv("../datasets/pokemon.csv")
sns.lmplot(x="Total", y="Defense", data=maindata, hue="Type 1" )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

maindata = pd.read_csv("../datasets/pokemon.csv")
sns.swarmplot(x="Type 1", y="Defense", data=maindata, hue="Type 1" )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

maindata = pd.read_csv("../datasets/mapsquito.csv")
sns.set(style="ticks")
sns.pairplot(data=maindata, hue="Type 1" )
plt.show()
```
